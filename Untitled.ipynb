```python
listing=['https://www.cian.ru/rent/flat/225415670',
'https://www.cian.ru/rent/flat/219238314',
'https://www.cian.ru/rent/flat/234164037',
'https://www.cian.ru/rent/flat/234955499',
'https://www.cian.ru/rent/flat/237446832',
'https://www.cian.ru/rent/flat/228235082',
'https://www.cian.ru/rent/flat/197351112',
'https://www.cian.ru/rent/flat/237773179',
'https://www.cian.ru/rent/flat/206616964',
'https://www.cian.ru/rent/flat/148535161',
'https://www.cian.ru/rent/flat/191502035',
'https://www.cian.ru/rent/flat/226755447',
'https://www.cian.ru/rent/flat/224329386',
'https://www.cian.ru/rent/flat/227982495',
'https://www.cian.ru/rent/flat/234238922',
'https://www.cian.ru/rent/flat/235624424',
'https://www.cian.ru/rent/flat/230951319',
'https://www.cian.ru/rent/flat/172840892',
'https://www.cian.ru/rent/flat/233896444',
'https://www.cian.ru/rent/flat/219360597',
'https://www.cian.ru/rent/flat/238607676',
'https://www.cian.ru/rent/flat/236583751',
'https://www.cian.ru/rent/flat/235640619',
'https://www.cian.ru/rent/flat/234081588',
'https://www.cian.ru/rent/flat/236238942',
'https://www.cian.ru/rent/flat/227073284',
'https://www.cian.ru/rent/flat/226629472',
'https://www.cian.ru/rent/flat/235397715',
'https://www.cian.ru/rent/flat/224990501',
'https://www.cian.ru/rent/flat/227491986',
'https://www.cian.ru/rent/flat/236790350']
l=[]
#f = open("/home/nbuser/library/ciandb.html", encoding="utf8") 
#soup = BeautifulSoup(f,"html.parser")
#f.close()
#containers=soup.find_all("div",{"class":"dece2e673c--container--79Dxy"})
#for container in containers:
    #print(str(container.find("a")).split('"')[1])
    #listing.append(str(container.find("a")).split('"')[1])
#print(listing)
for link in listing:
    print(link)
    time.sleep(5)
    r=requests.get(link, headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
    soup=BeautifulSoup(r.content,"html.parser")
    d={}
    #print(soup)
    if soup.find_all("div",{"data-name":"OfferUnpublished"}):
        d["Link"]=link
        d["Adress"]="СНЯТО С ПУБЛИКАЦИИ"    
    else:
        header=soup.find_all("div",{"class":"a10a3f92e9--header-information--38LX9"})
        pricetag=soup.find_all("div",{"class":"a10a3f92e9--price--1HD9F a10a3f92e9--price--residential--2ev_G"})
        coments=soup.find_all("div",{"class":"a10a3f92e9--info--10qpS"})
        generalinfo=soup.find_all("div",{"class":"a10a3f92e9--section_divider--1zGrv"})
        addfeatures=soup.find_all("li",{"data-name":"AdditionalFeatureItem"})
        
        d["Link"]=link
        d["dishwasher"]=0
        d["conditioner"]=0
        d["kids"]=0
        d["pets"]=0
        d["Adress"]=str(header[0].find("span",{"itemprop":"name"})).split('"')[1]
        d["Area"]=str(header[0].find("h1",{"data-name":"OfferTitle"})).split('"')[4].split(",")[1].replace("</h1>","").split(" ")[1]
        d["Price"]=str(pricetag[0].find("span",{"itemprop":"price"})).split('"')[1].replace(" ","").replace("₽/мес.","")
        d["Advance"]=str(coments[0].find("p",{"class":"a10a3f92e9--description--2xRVn"})).split(">")[1].split(",")[0].replace("Залог","").replace("₽","").replace("\xa0","")
        d["commision"]=str(coments[0].find("p",{"class":"a10a3f92e9--description--2xRVn"})).split(">")[1].split(",")[1].replace("комиссия","").replace("\xa0","")
        for item in addfeatures:
            if str(item.find("span",{"class":"a10a3f92e9--name--3bt8k"})).split(">")[1].replace("</span","")=="Санузел":
                if str(item.find("span",{"class":"a10a3f92e9--value--3Ftu5"})).split(">")[1].replace("</span","")=="1 раздельный":
                    wc=1
                elif str(item.find("span",{"class":"a10a3f92e9--value--3Ftu5"})).split(">")[1].replace("</span","")=="1 совмещенный, 1 раздельный":
                    wc=2
                elif str(item.find("span",{"class":"a10a3f92e9--value--3Ftu5"})).split(">")[1].replace("</span","")=="2 совмещенных":
                    wc=2
                else:
                    wc=0
        d["WC"]=wc

        if str(generalinfo[0].find("li",{"class":"a10a3f92e9--item--Ecr_d a10a3f92e9--dishwasher--2Jedr"}))!="None": d["dishwasher"]=1
        if str(generalinfo[0].find("li",{"class":"a10a3f92e9--item--Ecr_d a10a3f92e9--cond--3WZYH"}))!="None": d["conditioner"]=1
        if str(generalinfo[0].find("li",{"class":"a10a3f92e9--item--21VpQ a10a3f92e9--kids--3mK0H"}))!="None": d["kids"]=1
        if str(generalinfo[0].find("li",{"class":"a10a3f92e9--item--21VpQ a10a3f92e9--pets--3gLzA"}))!="None": d["pets"]=1
    l.append(d)
df=pd.DataFrame(l)
#сумма первого платежа аванс плюс комиссия плюс месячный платеж
#текст обьявления
#гардеробная 
#субьективные рейтинги модно проставлять после просмора, а просмотр по рейтингуделать
print("all")

df
```
